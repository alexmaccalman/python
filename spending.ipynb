```python
import pandas as pd
filepath = r"C:\gitrepos\bk_download.csv"
df = pd.read_csv(filepath)
df = df[df['Category'] != 'Category Pending'].reset_index()
```

```python
import re

def clean(string, remove):
    new = re.sub(rf"[{remove}]", "", string)
    new = " ".join(new.split())
    new = new.strip()
    return new
```

```python
categories = df['Category']
df['Category'] = df['Category'].apply(lambda row, :clean(row, remove="&")).values
```

```python
cats = df['Category'].value_counts()
grouped = df['Amount'].groupby(df['Category'])
grouped.agg('sum').sort_values()
```
