### Accessing Data with Pandas

```python
# single row
df.iloc[0] # pass in index
```

```python
df.loc[[0, 1, 2]] # access multiple indexes, rows
```

```python
df[:3] # slices 0 to 2 indexes
```

```python
df[3:6] # or other range 3 to 5
```

```python
df.id.head(2) # index a column by column name
df['id'].head(2) # or like this
```

```python
df[['id', 'age', 'gender']][:2] # double bracket returns a dataframe
```

```python
df.age.iloc[0] # returns a single index valure from a column
```

```python
df.age.iloc[[0]] # a duoble bracket has a diffrent return type
```

```python
(df.age == 20).head(2) # retruns a boolean series
```

```python
df[df.age == 20].head(2) # returns a subset of the data as a dataframe
```

```python
# ~ is no, | is or, & is and
df[(df.age == 20) & (df.weight > 60)]. head(2)
```

```python
df[df.device.str.split().apply(lambda x: len(x) == 1)].head(3) # adds a lambda function as filter criteria
```
